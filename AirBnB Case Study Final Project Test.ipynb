```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('files/AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?
#air_bnb.groupby('neighbourhood_group').head(5)
air_bnb.groupby('neighbourhood_group').count().mode()
# Manhattan
```

```python
# Are private rooms the most popular in manhattan?
air_bnb.groupby('neighbourhood_group')['room_type'].head()
# No!
```

```python
# Which hosts are the busiest and based on their reviews?
air_bnb.groupby('host_id')['number_of_reviews'].count().nlargest(5)
# All of these have 0 nights available
```

```python
#Which neighorhood group has the highest average price?
air_bnb.groupby('neighbourhood_group')['price'].mean().nlargest()
```

```python
# Which neighbor hood group has the highest total price?
air_bnb.groupby('neighbourhood_group')['price'].sum().nlargest()
```

```python
#Which top 5 hosts have the highest total price?
# air_bnb.groupby('host_name')['reviews_per_month'].sum().nlargest()
# air_bnb.groupby(['host_name', 'price'])['reviews_per_month'].sum().nlargest()
air_bnb.groupby(['host_name', 'price'])['reviews_per_month'].sum().nlargest()
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

# air_bnb.query(['number_of_reviews > 100','host_id'])['availability_365'].head()
air_bnb.query('number_of_reviews > 100')['availability_365'].head()
```

```python
# What host has the highest total of prices and where are they located?

air_bnb.groupby(['host_name', 'neighbourhood'])['price'].sum().nlargest(1)
```

```python
# When did Danielle from Queens last receive a review?


air_bnb.groupby('host_name','')['last_review'].head()
```
